## Supervisado, semisupervisado, no supervisada y aprendizaje reforzado 

Los algoritmos de ML son generalmente divididos en cuatro categorias basicas: **Supervisada**, **no-supervisada**, **semi supervisada** y de aprendizaje reforzado.

### Aprendizaje supervisado

Aprendizaje **supervisado** se refiere que el aprendizaje esta basado en datos **etiquetados**, y la construccion de los modelos es tal que uno quiere predecir a que etiqueta pertenece un conjunto de datos que no ha sido visto por el modelo anteriormente. 

ALgo asi como que uno tiene dos conjuntos: 

\begin{equation}
\left[\; \mathbf{datos\ entrada} \; + \; \mathbf{resultado\ correcto} \; \right] \rightarrow \mathbf{predecir\ resultados\ para\ nuevos\ datos}
\end{equation}

*Ejemplo:* Dado un conjunto de palabras en un cierto idioma, crear un algoritmo que sea capaz de predecir a que idioma pertenece.

### Aprendizaje no supervisado

Aprendizaje **no-supervisado** se refiere a datos que no esta **etiquedatos**, y el algoritmo tiene que encontrar la estructura de los datos para podemos agregarlos "por similitud". 

\begin{equation}
\left[\; \mathbf{datos\ entrada} \; + \; ? \; \right] \rightarrow \mathbf{encuentre\ estructura\ presente\ datos}
\end{equation}


*Ejemplo:* Dados unas fotografias, clasificar las fotograficas de acuerdo a las personas que aparecen en ellas.

### Aprendizaje semi supervisado

Aprendizaje **semi-supervisado** se refiere al caso donde tenemos un poco de datos que estan etiquetados y luego tenemos mas datos pero estos no estan etiquetados. 

*Ejemplo:* encontrar situaciones anomalas dentro de una red de trafico.

### Aprendizaje reforzado

Aprendizaje **reforzado** se refiere a situaciones donde tenemos muy poco conocimiento de como lucen los resultados esperados, pero si podemos es crear una funcion que califica que tan buena es una solucion dada basados en un objetivo final claro. 

\begin{equation}
\left[\; \mathbf{datos\ entrada} \; + \; \mathit{alguna\ salida} \; + \mathbf{funcion\ calificadora} \;\right] \rightarrow \mathbf{la\ mejor\ estrategia\ para\ lograr\ una\ solucion }
\end{equation}


*Example:* Los algoritmos usados en vehiculos autonomos o algoritmos que aprendje a jugar un juego de tablero.

# Clasificación and Regresion


En **aprendizaje supervisado**, lo que nosotros tenemos son datos que consisten tanto en caracteristicas, $\vec{X}$ y etiquetas, $\vec{Y}$. 

La idea es entonces construir un estimado que sea capaz de predecir la etiqueta de un objeto, $f{\vec{X})$, si las caracteristicas son conocidas. Por ejemplo predecir el precio de una casa, si sabemos que la casa esta localizada en cierta colonia, que tiene cierta area construida, que esta "cerca" de colegios, etc.  Tambien esta la de predecir si un usuario es un posible comprador basados en los movimientos que el hace en el mouse cuando visita una pagina Web, etc. 


Lo que tienen en comun estas tareas, es que hay uno o mas de cantidades desconocidas que estan asociades con el objeto que necesitan ser determinados desde cantidades conocidas o observadas.

Las aplicaciones en aprendizaje supervisado se puede dividir en dos grandes categorias: **clasificacion** and **regresion**. En clasificacion, la etiqueta es discreta, mientras que en regresion, la etiqueta es continua.

La idea de una tarea de clasificacion es predcir si una observacion data dentro de un conjunto de datos posee una propiedad o atributo en particular. Para realizar esas predicciones, nosotros medimos los atributos de varias observaciones que han sido etiquetadas y por lo tanto podemos comparar este conocimiento con observaciones que no han sido etiquetadas.

Mientras que el interes de la regresion es el predcir una variable usualmente continua. Supongo que ustedes ya usuaron alguna vez la regresion lineal para fitear una linea recta a unos datos. En este caso esta relacion puede derivarse matematicamente, pero el problema es cuando la dimensionalidad de los datos incrementa y por lo tanto ML es el metodo a usar.

### Clasificacion usando K-Nearest Neighbors

Con un clasificador [**K-Nearest Neighbors**](https://en.wikipedia.org/wiki/K-nearest_neighbors_algorithm), comen zamos con un conjunto de datos que han sido etiquetados (por ejemplo, los 200 libros que creamos antes y sus etiquetas, y luego agreramos es un conjunto de observaciones que no han sido etiquetadas. Para cada una de ellas, medimos que tan cercanos es a los datos ya clasificados y escogemos los "K" mas cercanos, donde K es un numero impar que usamos para la clasificacion. De este grupo encontramos las etiquetas mas comunes ("los vecinos K mas cercanos) y le datos esta etiqueta al datos de interes.

La siguiente grafica muestra el escenario. Una nueva observacion (el signo de pregunta) tiene algunos puntos cercanos a esta que son etiquetados por triangulos o estrella. Suponga que escogemos usar un K de valor 3. En este caso, consultamos los 3 puntops mas cercanos del simbolo de pregunta (en este caso, estrella, triangulo, triangulo). Usamos el voto de la mayoria y por lo tanto le damos la etiqueta de triangulo al simbolo de pregunta.

<img src='fig/knn.gif'>


SI hubieramos usado K=1, hubieramos escogido estrella, pero si escogemos K=3 o K=5, entonces es un triagulo. Por lo tanto el parametro K es un parametro importante del clasificador.

Un nota que nos servira mas adelante. Loa algoritmos pueden ser de aprendizaje ansios, que generan un modelo generalizado antes de hacer una prediccion. Son algoritmos que estan "listos" y activos para clasificar datos no clasificados. 

Tambien hay algoritmos perezosos, que no necesitan de aprendizaje o entrenamiento del modelo y que todos los puntos son usados en el momento de la prediccion. Ellos esperan hasta el ultimo minuto para hacer una clasificacion. Es decir que ellos gastan mas tiempo en la prediccion que en el entrenamiento.

Otra nota, pero esta vez refiriendonos en particular a KNN. KNN es mucho mejor cuando tenemos un numero "pequeño" de caracteristicas. Entre mas grande el numero de caracteristicas mas datos se necesitaran para predecir. Ademas si incrementamos la dimensionalidad, podemos tener el problema de sobreajuste (overfitting). Para evitar este problema, el numero de datos que necesitamos crece muy rapido (exponencialmente) a medida que incrementamos el numero de dimensiones. Este problema de la dimensionalidad se suele conocer en ingles como "curse de dimensionality" (la maldicion de la dimensionalidad).

Como veremos mas adelante, siempre que vayamos a hacer un modelo de ML, tenemos que reducir la dimension de las caracteristicas, por ejemplo usando un algoritmo del que hablaremos despues, que es el llamado "principal component analysis" o usar algun metodo para seleccionar las caracteristicas mas importantes. Es decir primer tienes que hacer PCA antes de usar cualquier algoritmo de ML. Otro punto es que cuando la dimension es muy grande, usar una distancia euclidiana ya no es tan eficiciente y se usan otras metricas ("cosine similarity"), pero no se preocupen, ya hablaremos mas de esto.

```python
# esta vez, yo voy a crear los datos

# Primera caracteristica
weather=['Sunny','Rainy','Rainy','Rainy','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny',
'Rainy','Sunny','Overcast','Overcast','Rainy']
# Segunda
temp=['Hot','Cool','Cool','Cool','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild']

#Variable que queremos predicir
play=['No','No','No','No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
```

La mayoria de algoritmos requieren es datos de entrada numericos, por lo que tenemos que cambiar las columnas que nos dan categorias en cada columna para que en vez nos den numeros. Esto se llama **Label Encoding** y sklearn tiene utilidades que hacen esto.

```python
# Importe LabelEncoder
from sklearn import preprocessing
#cree labelEncoder
le = preprocessing.LabelEncoder()
# Convierta las etiquetas de palabras en numeros. 
weather_encoded=le.fit_transform(weather)
print(weather_encoded)
```

```python
# Convierta las etiquetas de palabras en numeros.
temp_encoded=le.fit_transform(temp)
label=le.fit_transform(play)
```

```python
# Ahora vamos a combinar las diferentes propiedades en un sola tabla de "propiedades"
#
features=list(zip(weather_encoded,temp_encoded))
```

```python
# Ahora si usemos KNN
model = KNeighborsClassifier(n_neighbors=3)

# Train the model using the training sets
model.fit(features,label)

#Predict Output
predicted= model.predict([[0,2]]) # 0:Overcast, 2:Mild
print(predicted)
```

# Deteccion de Anomalias

[**Deteccion de Anomalias**](https://en.wikipedia.org/wiki/Anomaly_detection) se refiere a la identificacion de anomalias, o datos que estan fuera de lo normal. Encontrarlos en datos una dimension, es muy facil pero encontrarlos en problemas de gran dimensionalidad es mas dificil.

El interes de estos algoritmos va en una gran cantidad de dominios del conocimiento, desde detectar transacciones fraudulentas en un banco, deteccion de intrusos en una red de computo, incremento/reduccion inesperado en las ventas, cambio en el comportamiento de clientes, etc. 

Una de las tecnicas para clasificar anomalias en datos con gran dimensionalidad es el denominado
[**Isolation Forest**](https://cs.nju.edu.cn/zhouzh/zhouzh.files/publication/icdm08b.pdf). Este algoritmo de **bosque de aislamiento** identifica valores atipicos en un conjunto de datos dividiendo el espacio aleatoriamente hasta que cada punto es aislado de los demas. Repitiendo este procedimiento varias veces, este algoritmo identifica puntos que estan aislados de otros. El proceso es no supervisado, es decir que no tenemos que conocer las etiquetas que definen un dato aislado. Una buena explicacion la pueden encontrar [aqui](https://towardsdatascience.com/how-to-perform-anomaly-detection-with-the-isolation-forest-algorithm-e8c8372520bc), de donde extraemos la imagen:



Isolation Forest no son mas que un conjunto de arboles de decision binarios (si o no). Cada arbol en este algoritmo es llamada un iTree. El algoritmo comienza siendo entrenado por los datos que ya tienen , generando entonces "arboles aislados". 

Pero para entenderlo bien, lo mejor sera explicar otro metodo que es mas general y muy usado, el denominado "random forest". No es mas que un conjunto de arboles de decision (decision trees), pueden ser usados para clasificacion y para regresion (de nuevo esto es solo un apetizer, veremos mas adelante esto con mas detalle). La idea es dividir los datos en ramas de los arboles y al final tomar decisiones basados en un metodo de mayoria o tomar el promedio si lo que queremos es predecir un valor. 

Como funciona?

Ahora entrenamos un arbol de decision para cada conjunto de datos, independientemente. Pero hay un detalle, no usaremos todas las caracteristicas de los datos, vamos a seleccionar aleatoriamente dentro las caracteristicas de los datos. Por ejemplo para *ID1* usaremos *X1* y *X2*, para *ID2* usaremos *X2* y *X4*, *ID3* usaremos *X1* y *X3* y *ID4* usaremos *X3* y *X4*


Y ahora sencillamente lo que hacemos es hacer el siguiente proceso para cada ID.

Generamos condiciones que deben cumplir, por ejemplo x1<3 o cosas asi (encuentre el min y el max de la variable x1 y escoja aleatoriamente cual es el valor a comparar. 

Ok, en principio todo parece muy simple, pero no es tan asi, dependiendo de los datos, es posible que cuando hagamos el arbol no lleguemos a tener un arbol que haga que las hojas del arbol tengan elementos que solo pertenecen a esa clase. Entonces como sabemos cuando un "split" es mejor que otro? es decir cuando la maquina va a escoger el mejor proceso?. Para esto usamos una definicion creada en la teoria de la informacion, que es medir una cantidad que se llama ganancia de informacion (information gain) o para los fisicos, la entropia.

Inicialmente tenemos 6 valores y 3 de ellos son 0 y 3 son 1. Es decir que si nosotros eligieramos aleatoriamente uno, solo acertariamos la mitad de las veces. Lo que queremos es hacer eso y para eso lo que hacemos es que la informacion que ganemos al hacer el arbol aumente lo mas posible. Matematicamente, esto requiere que maximizemos la cantidad llamada entropia

$S= - \sum_i p_i \ln(p_i)$

donde $p_i$ es la probabilidad de estar en la clase $p_i$. Por ejemplo para nuestro caso inicial, donde teniamos una probabilidad de 0.5 para cada caso entonces $S = - 2*(0.5 \log_2 0.5)=1.0$, que es el valor mas grande de la entropia. Ahora por ejemplo, si medimos la entropia en una hoja del arbol donde solo hay un tipo de ellos $-1*\log_2(1)-0*\log_2(0)=0$

Ok, ya tenemos como medir la "entropia en cada una de las ramas del arbol, pero lo que nos interesa es saber cuanta entropia ganamos a medida que nos vamos moviendo en el arbol. Para un split en particular, lo unico que hacemos es substraer la entropia de los hijos (luego del split) a la del papa, como si fuera

$EG=S(\text{papa})-\sum_i \omega_i S(\text{hijo}_i)$, donde $\omega_i$ es el tamaño relativo del hijo con respecto al papa (cuantos datos pasaron a esa "hoja" del total). Por lo tanto la calidad de un arbol es medida con respecto al "information gain".

# Reduccion de la dimensionalidad

En uno de los primeros ejemplos que discutimos, teniamos solo 4 etiquetas (el tipo de libro) y usamos esas dos para clasificar un nuevo libro bastado en las veces que una palabra aparecia en el nuevo libro. Aunque solo tiene 4 etiquetas, el poder visualizar esto es bastante mas complicado. En escenarios mas reales, el numero de etiquetas pueden ser decenas, centenas e inclusive miles. En nuestro caso podriamos usar mas palabras, pero no porque aumentemos el numero de palabras, garantizamos que vamos a hacer una mejor clasificacion. Seguro que hay palabras que no son estrictamente usadas en una sola disciplina, por lo que dficultara la clasificacion en general.


Datos en altas dimensiones puedes ser dificil de trabaja y poseen problemas especificos para los modelos de ML ( vea **[The Curse of Dimensionality](https://en.wikipedia.org/wiki/Curse_of_dimensionality)**). Por lo tanto lo ideal seria tener una tecnica que nos ayude a recudir la dimensionalidad del espacio y nos ayude a enfocarnos unicamente en las etiquetas que sean mas relevantes.


Una tecnica muy popular para reducir la dimensionalidad de un conjunto de datos es usar una tecnica llamada **[Analisis de componentes principales](https://en.wikipedia.org/wiki/Principal_component_analysis)** (PCA por sus siglas en ingles). PCA trata de encontrar una representacion de dimensionalidad mas baja del conjunto de datos proyectando los datos a un espacio de dimensionalidad mas pequeña y haciendo que haya un costo muy pequeño  a la perdida de informacion.

Para tener una intuicion acerca de PCA, suponga que tenemos puntos en dos dimensiones y lo que queremos es reducir los puntos a una dimension. Para hacer eso, tu puedes encontrar el centro de los puntos y crear una linea $L$ con una orientacion aleatoria que pasa a traves de ese punto (el centro de masa). Y ahora lo que podemos hacer es proyectar cada punto sobre la linea. Cada punto entonces se representa sobre una posicion dentro de la linea. Y podemos usar la diferencia entre el puntos mas lejanos del centro de masa como una medida de la "varianza de los datos", entre mas grande sea esta valor entonces mas "cercano" es a la representacion original.

Lo que puede uno hacer es rotar la linea y media este factor que llamamos delta, y por lo tanto encontramos la orientacion de la linea donde la perdida de informacion se disminuye.

# Agregamiento

Agremaniento de datos es una tecnica muy poderosa de ML, en la cual uno tiene que definir de alguna manera una distancia (en matematicas dirian que hay que definir metrica porque tiene que tener varias propiedades). Al final, la meta de los algoritmos de agregamiento es crear grupos de observaciones, donde cada grupo tiene observaciones similares.

En esta area hay un conjunto importante de metodos, incluyendo [density-based clustering](https://en.wikipedia.org/wiki/DBSCAN), [hierarchical clustering](https://en.wikipedia.org/wiki/Hierarchical_clustering), y [centroid clustering] Uno de los metodos mas simples y mas usados es el denominado [**agregamienti K-Means**](https://en.wikipedia.org/wiki/K-means_clustering). Dada una coleccion de datos en el espacio, K-means selecciona *k* puntos centroides, primero generados aleatoriamente (en verde abajo), y asigna a cada punto de los datos un centroide debido a su cercania. Esando estos grupos preliminares, lo siguiente es calcular el centroide cada grupo (centro de masa)... o matematicamente
$\sum \vec{r}_i/N$. Estos nuevos seran los nuevos centroides y uno vuelve a asignar los centroides a los puntos y repite este proceso hasta que los centroides no cambien. Un realizacion visual de este algoritmo se puede ver aqui:

Veamos ahrcomo lo podemos usar en una situacion mas compleja y con dimensiones mas altas. Para este ejemplo, vamos a usar datos que vienen de la universidad de Stanford y que son usados en NLP, [**GloVe vectors**](https://nlp.stanford.edu/projects/glove/), que lo que hacen es que dado un "corpus" de palabras crea vectores de numeros reales que representan estas palabras. Para esto vamos a leer el achivo GloVe, extraer las primeras `n` palabras y sus correspondiente 50 vectores, y vamos a examinar el contenido de la primera palabra.

```python
from zipfile import ZipFile
from collections import defaultdict
from urllib.request import urlretrieve
import numpy as np
import json, os, codecs

# download the vector files we'll use
if not os.path.exists('glove.6B.50d.txt'):
  urlretrieve('http://nlp.stanford.edu/data/glove.6B.zip', 'glove.6B.zip')
  # unzip the downloaded zip archive
  ZipFile('glove.6B.zip').extractall(os.getcwd())

# get the first n words and their vectors
vectors = []
words = []
n = 50000
for row_idx, row in enumerate(codecs.open('glove.6B.50d.txt', 'r', 'utf8')):
  if row_idx > n: break
  split_row = row.split()
  word, vector = ' '.join(split_row[:-50]), [float(i) for i in split_row[-50:]]
  words += [word]
  vectors += [vector]
    
# check out a sample word and its vector
print(words[1700], vectors[1700], '\n')
```

como pueden ver, <code>words</code> es solo una lista de palabras. Para cada una de palabras, hay un vector que es de 50 dimensiones. En principio, cada vector abstrae del significado semantico de la palabra. Es decir si el lenguaje español fuerea in espacio de 50 dimensiones, cada palabra del idioma tendria un punto en este espacio multidimensional. Por lo tanto cada palabra en <code>words</code> estaria dentro de espacio y representada por su vector.

Palabras que tienen un significado similar puede aparecer muy cercanas las unas de las otras dentro de este espacio de vectores. Para probar este hipotesis, vamos a usar agregamiento K-mean para identificar 20 agregados (clusters) dentro de ese espacio de 50 dimensiones. Luego, vamos a crear un mapa que llamaremos 
<code>groups</code> donde las "claves" seran los clusters que tienen identificacion (0-19) y cuyos valores seran las listas de palabras que pertenecen a ese cluster. Luego de crear esta variable, vamos a imprimir las primeras 10 palabras en cada cluster:

```python
from sklearn.cluster import KMeans

# cluster the word vectors
kmeans = KMeans(n_clusters=20, random_state=0).fit(np.array(vectors))

# `kmeans.labels_` is an array whos `i-th` member identifies the group to which
# the `i-th` word in `words` is assigned
groups = defaultdict(list)
for idx, i in enumerate(kmeans.labels_):
  groups[i] += [words[idx]]

# print the top 10 words contained in each group
for i in groups:
  print(groups[i][:10])
```
